```python
df = pd.read_csv("Crawling_data_쿠팡이츠2.csv")
df['DATE'] = pd.to_datetime(df['DATE'].astype(str), format='%Y-%m-%d')
mask = (df['DATE'] > '2019-06-01') & (df['DATE'] <= '2021-07-31')
df = df.loc[mask]
df = df.iloc[:,[1,3]].reset_index(drop=False, inplace=False)
df = df.iloc[:,[1,2]]
df.head()
```

#### STAR 열의 문자열을 별점수로 표시

```python
for i in range(len(df['STAR'])):
    a = re.findall("\d", df['STAR'][i])
    df['STAR'][i] = int(a[1])

df = df.astype({'STAR':'int'})
df.head()
```

#### 한국어 불용어 사전을 사용하고 text_cleansing함수를 정의

```python
stopwords = pd.read_csv("stopword.txt")
with open('stopword.txt', 'r', encoding='utf-8-sig') as f:
    data = f.read()
stopword = data.splitlines()
stopword
```

```python
def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 정규 표현식 처리
    result = hangul.sub('', text)
    okt = Okt()  # 형태소 추출
    nouns = okt.nouns(result)
    nouns = [x for x in nouns if len(x) > 1]  # 한글자 키워드 제거
    nouns = [x for x in nouns if x not in stopwords]  # 불용어 제거
    return nouns
```

### 임베딩

#### BoW 벡터 생성

```python
vect = CountVectorizer(tokenizer = lambda x: text_cleaning(x)) # tokenizer로 위에서 정의한 text_cleansing 함수 사용
bow_vect = vect.fit_transform(df['REVIEW'].tolist())
word_list = vect.get_feature_names()
count_list = bow_vect.toarray().sum(axis=0)
```

#### TF-IDF 적용

```python
tfidf_vectorizer = TfidfTransformer()
tf_idf_vect = tfidf_vectorizer.fit_transform(bow_vect)
invert_index_vectorizer = {v : k for k, v in vect.vocabulary_.items()}
print(str(invert_index_vectorizer)[:100]+'...')
```

#### 별점 라벨링(1,2,3점은 부정 4,5점은 긍정)

```python
def rating_to_label(rating):
    if rating > 3:
        return 1
    else:
        return 0
    
df['target'] = df['STAR'].apply(lambda x: rating_to_label(x))
```

```python
# 텍스트를 한 줄씩 처리
from konlpy.tag import Okt

okt = Okt()
results = []
lines = df['REVIEW'].tolist()

for line in lines:
    # 단어의 기본형 사용
    malist = okt.nouns(line) # 토큰화
    malist = [x for x in malist if len(x) > 1]  # 한글자 키워드 제거
    malist = [x for x in malist if x not in stopword] # 불용어 제거
    results.append(malist)
```

## 토픽모델링

### 전처리 및 토큰화

#### 4,5점을 제외한 리뷰데이터만 추출

```python
df_45x = []
for i in range(len(df)):
    if df['STAR'][i] < 4:
        df_45x.append(df['REVIEW'][i])
    else:
        continue
        
df_45x = pd.DataFrame(df_45x)
df_45x.rename(columns={0:'REVIEW'}, inplace=True)
```

#### apply_regular_expression 함수 정의

```python
def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

#### 정규 표현식 적용

```python
for i in range(len(df_45x)):
    df_45x['REVIEW'][i] = apply_regular_expression(df_45x['REVIEW'][i])
```

#### 토큰화

```python
okt = Okt()
docs = []
lines = df_45x['REVIEW'].tolist()

for line in lines:
    # 단어의 기본형 사용
    malist = okt.nouns(line) # 토큰화
    malist = [x for x in malist if len(x) > 1]  # 한글자 키워드 제거
    malist = [x for x in malist if x not in stopword]
    docs.append(malist)
```

#### get_nouns 함수 정의

```python
def get_nouns(text):
    nouns=okt.nouns(text)
    nouns=[word for word in nouns if len(word)>1]
    nouns = [word for word in nouns if word not in stopword]
    return nouns
```
